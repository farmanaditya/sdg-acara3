# 9 Mengkonstruksi Data Bagian 1

### Ekstraksi latitude & longitude berdasarkan fitur 'location'

```python
df.columns
```

```python
df.head(2)
```

```python
rest_loc = pd.DataFrame() ## Membuat dataframe untuk semua Location yang unik di Bangalore
```

```python
rest_loc['Name'] = df['location'].unique()
```

```python
rest_loc
```

```python
#Nominatim merupakan tool untuk mencari data pada OpenStreetMap berdasarkan alamat atau lokasi.
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent='app' , timeout=None)
```

```python
lat = [] ## List untuk menyimpan data Latitude
lon = [] ## List untuk menyimpan data Longitude

for name in rest_loc['Name']:
    location = geolocator.geocode(name)
    
    if location is None:
        lat.append(np.nan)
        lon.append(np.nan)
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)
```

```python
print(lat) 
```

```python
print(lon)
```

```python
rest_loc['lat'] = lat
rest_loc['lon'] = lon
```

```python
rest_loc
```

```python
## Berhasil mendapatkan masing-masing koordinat Lat dan Lon pada dataset menggunakan Geopy
```
